```python
X = df.drop('quality', axis=1)
y = df['quality']
print(X.head(5))
print (y.head(5))


```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential  # 'Sequential' should be capitalized

# Define the model
model = Sequential()  # 'Sequential' should have a capital 'S'
model.add(layers.Dense(11, activation="relu", input_dim=11))
model.add(layers.Dense(5, activation="relu"))
model.add(layers.Dense(10, activation="softmax"))  # Ensure output layer has the correct number of units


```

```python
# Compile and fit the model
model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])

history = model.fit(X, y, validation_split=0.22, batch_size=100, epochs=20)

model.summary()

```

```python
# Plot graph
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

```python
# Plot graph
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

```python
# MAKE PREDICTIONS WITH THE TRAINED MODEL

import numpy as np
from numpy import array

# input for 11 features
# Xnew = np.array([[7.9, 0.6, 0.06, 1.6, 0.069, 15.0, 59.0, 0.9964, 3.3, 0.46, 9.4]])
Xnew = np.array([[7.6, 0.31, 0.34, 2.5, 0.081, 26.0, 35.0, 0.99356, 3.22, 0.59, 12.5]])

# convert into numpy array
Xnew = np.array(Xnew, dtype=np.float64)

# making the prediction
Ynew = model.predict(Xnew)

print(Ynew)

max_index_row = np.argmax(Ynew, axis=1)

print(max_index_row)
```
